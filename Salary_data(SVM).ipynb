```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import svm 
from sklearn.svm import SVC
from sklearn import preprocessing 
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.metrics import classification_report
```

```python
sd_train=pd.read_csv("SalaryData_Train(1).csv")
sd_train.head()
```

```python
sd_test=pd.read_csv("SalaryData_Test(1).csv")
sd_test.head()
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
```

```python
string_col=['workclass','education','maritalstatus','occupation','relationship','race','sex','native','Salary']
for i in string_col:
    sd_train[i]=label_encoder.fit_transform(sd_train[i])
    sd_test[i]=label_encoder.fit_transform(sd_test[i])
```

```python
sd_train
```

```python
sd_test
```

```python

train_x=sd_train.iloc[0:500,0:13]
train_y=sd_test.iloc[0:500,13]
test_x=sd_test.iloc[0:300,0:13]
test_y=sd_test.iloc[0:300,13]
```

```python
clf=SVC()
param_grid=[{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001]}]
gsv=GridSearchCV(clf,param_grid,cv=10)
gsv.fit(train_x,train_y)
```

```python
gsv.best_params_,gsv.best_score_
```

```python
clf=SVC(C=15,gamma=0.5)
clf.fit(train_x,train_y)
train_y_pred=clf.predict(train_x)
train_acc=accuracy_score(train_y,train_y_pred)
print("Train Accuracy=",train_acc)
test_y_pred=clf.predict(test_x)
test_acc=accuracy_score(test_y,test_y_pred)
print("Test Accuracy=",test_acc)
confusion_matrix(train_y,train_y_pred)
```

```python
clf=SVC(kernel="linear")
clf.fit(train_x,train_y)
train_y_pred=clf.predict(train_x)
train_acc=accuracy_score(train_y,train_y_pred)
print("Train Accuracy=",train_acc)
test_y_pred=clf.predict(test_x)
test_acc=accuracy_score(test_y,test_y_pred)
print("Test Accuracy=",test_acc)
```

```python
clf=SVC(kernel="poly")
clf.fit(train_x,train_y)
train_y_pred=clf.predict(train_x)
train_acc=accuracy_score(train_y,train_y_pred)
print("Train Accuracy=",train_acc)
test_y_pred=clf.predict(test_x)
test_acc=accuracy_score(test_y,test_y_pred)
print("Test Accuracy=",test_acc)
```
